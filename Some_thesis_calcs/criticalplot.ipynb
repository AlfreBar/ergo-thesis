```python
import numpy as np
import matplotlib.pyplot as plt
%config IPCompleter.greedy=True

```

```python
theta = np.linspace(0, np.pi/2, 100)
x1 = np.cos(theta)
x2 = np.sin(theta)
```

```python
fig,ax=plt.subplots(figsize=[15,10])
plt.axhline(y=0,xmax=0.5,linewidth=4, color='r',label="Critical Points")
plt.axhline(y=1,linewidth=5, color='g',label="Ising",ls="-.")
plt.axvline(x=1,linewidth=5, color='r')
plt.xlabel(r"$\lambda$",fontsize=30)
plt.ylabel(r"$\gamma$",fontsize=30)
plt.plot(x1,x2,ls=":",color='b',label="Factorization line",linewidth=5)
plt.ylim(0,1.5)
plt.xlim(0,2)
plt.text(0.3, 0.5, "Ordered", family="serif",fontsize=30)
plt.text(1.3, 0.5, "Disordered", family="serif",fontsize=30)
plt.tight_layout()
plt.legend()
plt.savefig("xymodelphaseplot")
```
