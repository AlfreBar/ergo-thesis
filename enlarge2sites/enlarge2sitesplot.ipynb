```python
from infinitetrial import infinite_system_algorithm,Block,H2
import numpy as np
from scipy.sparse import kron,identity
from numpy.linalg import eigvalsh
import matplotlib.pyplot as plt
```

```python
model_d = 2  # single-site basis size


sX = np.array([[0, 1], [1, 0]], dtype='complex') 
sY = np.array([[0, -1j], [1j, 0]], dtype="complex") 
Id = np.array([[1, 0], [0, 1]], dtype='complex')

lambdarange=np.linspace(0,2,20)

rhom={}
for l in lambdarange:

    H1 = -l*np.array([[1, 0], [0, -1]], dtype='d') #-h*np.array([[0, 1], [1, 0]], dtype='d') # single-site portion of H 

    site = Block(length=2, basis_size=4, operator_dict={
        "H": kron(H1,H1),
        "conn_Sx": kron(identity(2),sX),
        "conn_Sy": kron(identity(2),sY),
        })

    rhom[l]=infinite_system_algorithm(site,site, 10, 10,H2)
```

```python
eigs=[eigvalsh(rhom[x]) for x in lambdarange]
plt.plot(lambdarange,eigs)
plt.show()
```
