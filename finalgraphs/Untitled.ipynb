```python
import numpy as np
import matplotlib.pyplot as plt
from coefficientfunc import gfunc
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
L=1
lambdarange=np.linspace(0.1,2,500)
```

```python
lambdarange=np.linspace(0,2)
gammarange=np.linspace(0,1.5)
Xx, Yy = np.meshgrid(lambdarange,gammarange)

z=np.zeros((len(lambdarange),len(gammarange)))

for b,gs in enumerate(gammarange):

    for a, lam in enumerate(lambdarange):
        
        z[b,a]=gfunc(0,lam,gs)
        
fig = plt.figure(figsize=[10,10])
ax = fig.gca(projection='3d')

surf = ax.plot_surface(Xx, Yy, z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
ax.view_init(elev=30, azim=250)
ax.set_xlabel(r"$\lambda  \ (lambda)$",fontsize=15)
ax.set_ylabel(r"$\gamma \ (gamma) $",fontsize=15)
ax.set_zlabel(r"Max-Erg",fontsize=15)
plt.suptitle("L=1, Analitical",fontsize=15)
plt.savefig("3dplot_g0")
plt.show()
```

```python
lambdarange=np.linspace(0,2)
gammarange=np.linspace(0,1.5)
Xx, Yy = np.meshgrid(lambdarange,gammarange)

maxerg=np.zeros((len(lambdarange),len(gammarange)))

for b,gs in enumerate(gammarange):

    for a, lam in enumerate(lambdarange):
        
        maxerg[b,a]=gfunc(0,lam,gs)
    

z=np.gradient(maxerg,axis=0)
fig = plt.figure(figsize=[10,10])
ax = fig.gca(projection='3d')

surf = ax.plot_surface(Xx, Yy, z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
ax.view_init(elev=10, azim=240)
ax.set_xlabel(r"$\lambda  \ (lambda)$",fontsize=15)
ax.set_ylabel(r"$\gamma \ (gamma) $",fontsize=15)
ax.set_zlabel(r"first-der",fontsize=15)
plt.suptitle("L=1, Analitical",fontsize=15)
plt.savefig("3dplot_firstder")
plt.show()
```

```python
lambdarange=np.linspace(0,2,200)
gammarange=np.linspace(0,1.5,200)
Xx, Yy = np.meshgrid(lambdarange,gammarange)

maxerg=np.zeros((len(lambdarange),len(gammarange)))

for b,gs in enumerate(gammarange):

    for a, lam in enumerate(lambdarange):
        
        maxerg[b,a]=gfunc(0,lam,gs)
    

z=np.gradient(maxerg,axis=0)
f, ax = plt.subplots(1,figsize=(15,15))
im = ax.imshow(z)
xticks=np.array([0,1,2])
yticks=np.array([0,0.5,1,1.5])
ax.set_xticks(np.linspace(0,len(lambdarange)-1,len(xticks)))
ax.set_yticks(np.linspace(0,len(gammarange)-1,len(yticks)))
ax.set_xticklabels(xticks)
ax.set_yticklabels(yticks)
ax.set_xlabel(r"$\lambda  \ (lambda)$",fontsize=15)
ax.set_ylabel(r"$\gamma \ (gamma) $",fontsize=15)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.suptitle("L=1, Analitical First-Der",fontsize=15)

plt.colorbar(im, cax=cax)
plt.savefig("planarplot_firstder_onesite")

plt.show()
```
