```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def trial_fun(lam,phi):
    f1=(np.cos(phi/2))**2-(np.sin(phi/2))**2
    
    f2=(lam)*np.cos(phi/2)*np.sin(phi/2)
    
    return -(1/2)*f1**2-f2
```

```python
phi_range=np.linspace(0,np.pi)
for par in [0,0.5,1,1.5,2]:
    plt.plot(phi_range,trial_fun(par,phi_range),label=par)
    plt.vlines(np.pi/2,-1,0,linestyles="dashed")
    plt.hlines(-0.5,0,np.pi,linestyles="dashed")
    plt.legend()
    plt.xticks([0,np.pi/2,np.pi],[0,r"$\pi /2 $",r"$\pi$"])
```
